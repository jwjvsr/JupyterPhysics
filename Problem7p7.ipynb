Feedback system

##### Initialization

```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

##### Define the transfer functions for the open-loop and closed loop

Open-loop:
$$
H_{ol} = \frac{1}{s^2 + 4 s + 3}
$$
Closed-loop:
$$
H_{cl} = \frac{s+1}{s^2 + 4 s + 4}
$$

```python
sysol = signal.TransferFunction([1], [1, 4, 3])
syscl = signal.TransferFunction([1, 1],[1, 4, 4])
```

Calculate the poles and zero's.

```python
zpkol = signal.tf2zpk([1], [1, 4, 3])
zpkcl = signal.tf2zpk([1, 1],[1, 4, 4])
print(zpkol)
print(zpkcl)
```

Open-loop:
* poles: -3 and -1
* zero's: none

Closed-loop:
* poles: -2 and -2
* zero's: -1

##### Make Bode plots of the two transfer funcitons:

```python
ws = np.linspace(10**-1, 10**2, 600)
wcl, magcl, phasecl = syscl.bode(ws)
wol, magol, phaseol = sysol.bode(ws)
```

```python
plt.figure()
plt.semilogx(wcl, magcl)    # Bode magnitude plot system 1
plt.semilogx(wol, magol)    # Bode magnitude plot system 3
plt.grid(True)
plt.figure()
plt.semilogx(wcl, phasecl)  # Bode phase plot system 1
plt.semilogx(wol, phaseol)  # Bode phase plot system 3
plt.grid(True)
plt.show()
```
