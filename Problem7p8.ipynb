```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

##### First order system:

The first system is a first-order system that is defined via the transfer function model.
The model is:
$$
H_1(s) = \frac{1}{s+1}
$$
A Bode-plot is generated to analyse the frequency response.

Define the first-order tranfer function and the cascade of three first-order ones.
The transfer function of the cascaded system is:
$$
H_3(s) = \frac{1}{s^3 + 3 s^2 + 3 s + 1}
$$

```python
sys1 = signal.TransferFunction([1], [1, 1])
sys3 = signal.TransferFunction([1],[1, 3, 3, 1])
```

```python
w1, mag1, phase1 = sys1.bode()
w3, mag3, phase3 = sys3.bode()
```

```python
plt.figure()
plt.semilogx(w1, mag1)    # Bode magnitude plot system 1
plt.semilogx(w3, mag3)    # Bode magnitude plot system 3
plt.grid(True)
plt.figure()
plt.semilogx(w1, phase1)  # Bode phase plot system 1
plt.semilogx(w3, phase3)  # Bode phase plot system 3
plt.grid(True)
plt.show()
```

#### Calculate the step response of the two systems

```python
t1 = np.linspace(0, 10)
u1 = np.ones_like(t1)
tout1, y1, x1 = signal.lsim(sys1, u1, t1)
plt.plot(t1, y1)
plt.grid(True)
tout3, y3, x3 = signal.lsim(sys3, u1, t1)
plt.plot(tout3,y3)
plt.grid(True)
plt.show()
```

Define the value for the gain for later calculations and the closed-loop transfer function:
$$
H_{cl}(s) = \frac{k}{s^3 + 3 s^2 + 3 s + 1 + k}
$$

```python
k=2
syscl = signal.TransferFunction([k],[1,3,3,1+k])
wcl, magcl, phasecl = syscl.bode()
```

```python
plt.figure()
plt.semilogx(w3, mag3)    # Bode magnitude plot system 1
plt.semilogx(wcl, magcl)    # Bode magnitude plot system 3
plt.grid(True)
plt.figure()
plt.semilogx(w3, phase3)  # Bode phase plot system 1
plt.semilogx(wcl, phasecl)  # Bode phase plot system 3
plt.grid(True)
plt.show()
```

```python
tcl = np.linspace(0, 10)
ucl = np.ones_like(tcl)
toutcl, ycl, xcl = signal.lsim(syscl, ucl, tcl)
plt.plot(tcl, ycl)
plt.grid(True)
tout3, y3, x3 = signal.lsim(sys3, ucl, tcl)
plt.plot(tout3,y3)
plt.grid(True)
plt.show()
```
